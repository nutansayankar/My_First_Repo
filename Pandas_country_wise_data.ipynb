```python
summary=df.groupby('Country')[['New cases','New deaths','New recovered']].sum().reset_index()
summary
```

VISUALIZATION

Daily New COVID Cases

```python
plt.figure(figsize=(12,6))
sns.barplot(data=summary.nlargest(10,'New cases'),x='Country',y='New cases',palette='bright',hue='Country')
plt.title('Top 10 countries by Daily New COVID Cases')
plt.xticks(rotation=30)
plt.show()
```

The visualization shows that the highest number of new daily COVID-19 cases are recorded in the United States followed by Indian and Brazil.

Daily COVID Deaths

```python
plt.figure(figsize=(12,6))
sns.barplot(data=summary.nlargest(10,'New deaths'),x='Country',y='New deaths',palette='dark',hue='Country')
plt.title('Top 10 countries by Daily COVID Deaths')
plt.xticks(rotation=30)
plt.show()
```

The visualization shows that the highest number of daily COVID-19 deaths are recovered in the United States, followed by Indian and Brazil.

Daily Recovery Trend

```python
plt.figure(figsize=(12,6))
sns.barplot(data=summary.nlargest(10,'New recovered'),x='Country',y='New recovered',palette='colorblind',hue='Country')
plt.title('Top 10 countries by Daily COVID Deaths')
plt.xticks(rotation=30)
plt.show()
```

The visualization shows that the highest number of COVID-19 recoveries are recovered in Brazil followed by India and US
